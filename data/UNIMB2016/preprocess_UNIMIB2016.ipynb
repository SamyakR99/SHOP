### Get bounding boxes

```python
import json

# Specify the file path
file_path = '/home/samyakr2/food_datasets/UNIMIB2016/annotations/annotations.json'

# Open the JSON file for reading
with open(file_path, 'r') as file:
    # Load the JSON data
    data = json.load(file)
```

```python
print(data[1].keys())
print(len(data))
print(data[1]['objects'])
print(len(data[1]['objects']))
```

```python
from skimage.draw import polygon
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

segments_dict = {}

for idx in range (len(data)):
    if data[idx]['filename'] != '20151211_142219':
    
        img_path = '/home/samyakr2/food_datasets/UNIMIB2016/images/original/'+str(data[idx]['filename'])+'.jpg'
        im = Image.open(img_path)
        im_np = np.asarray(im, dtype=np.uint8)

        if not isinstance(data[idx]['objects'], dict):
            for num_seg in range (len(data[idx]['objects'])):
                file_name = data[idx]['filename']+'_'+str(num_seg)
                rr, cc = polygon(data[idx]['objects'][num_seg]['polygon_y'], data[idx]['objects'][num_seg]['polygon_x'])
                image = np.zeros(im_np.shape, dtype=np.uint8)
                rr = np.clip(rr, a_min=0, a_max=2447)
                cc = np.clip(cc, a_min=0, a_max=3263)

                image[rr, cc] = 1

                img = im_np * image
                non_dark_region = (img > 1).any(axis=-1)

                # Get the bounding box of the non-dark region
                non_dark_bbox = np.argwhere(non_dark_region)
                min_y, min_x = np.min(non_dark_bbox, axis=0)
                max_y, max_x = np.max(non_dark_bbox, axis=0)

                cropped_img = img[min_y:max_y, min_x:max_x, :]

                segments_dict[file_name] = cropped_img

        if isinstance(data[idx]['objects'], dict):
            file_name = data[idx]['filename']+'_0'
            rr, cc = polygon(data[idx]['objects']['polygon_y'], data[idx]['objects']['polygon_x'])
            image = np.zeros(im_np.shape, dtype=np.uint8)
            rr = np.clip(rr, a_min=0, a_max=2447)
            cc = np.clip(cc, a_min=0, a_max=3263)

            image[rr, cc] = 1

            img = im_np * image
            non_dark_region = (img > 1).any(axis=-1)

            # Get the bounding box of the non-dark region
            non_dark_bbox = np.argwhere(non_dark_region)
            min_y, min_x = np.min(non_dark_bbox, axis=0)
            max_y, max_x = np.max(non_dark_bbox, axis=0)

            cropped_img = img[min_y:max_y, min_x:max_x, :]

            segments_dict[file_name] = cropped_img
            
np.save('/home/samyakr2/food_datasets/UNIMIB2016/images/local_imgs.npy', segments_dict) 
```

```python
labels_list = []
for idx in range (len(data)):
    if not isinstance(data[idx]['objects'],dict):
        for num_seg in range (len(data[idx]['objects'])):
            if not data[idx]['objects'][num_seg]['type'] in labels_list:
                labels_list.append(data[idx]['objects'][num_seg]['type'])
    if isinstance(data[idx]['objects'],dict):
        if not data[idx]['objects']['type'] in labels_list:
                labels_list.append(data[idx]['objects']['type'])
```

```python
import csv
total_labels = []
with open('/home/samyakr2/food_datasets/UNIMIB2016/annotations/annotation.csv', 'r', newline='') as file:
    csv_reader = csv.reader(file)
    
    for row in csv_reader:
        items = str(row[1])
        if '[' in items:
            # print(items.split(','))
            ite = items.replace('[', '')
            it = ite.replace(']', '')
            
            it_split = it.split(',')
            for ele in it_split:
                if ele not in total_labels:
                    total_labels.append(ele)
```

```python
import numpy as np

train_dict = {}

with open('/home/samyakr2/food_datasets/UNIMIB2016/annotations/train_data.txt', 'r') as file:
    for line in file:
        id_name = line.split('/')[-1][:-5]
        
        with open('/home/samyakr2/food_datasets/UNIMIB2016/annotations/annotation.csv', 'r', newline='') as file:
            csv_reader = csv.reader(file)

            for row in csv_reader:
                if id_name in row[0]:
                    multi_labels = [1 if label in row[1] else 0 for label in final_label]
                    train_dict[id_name] = multi_labels

np.save('UNIMIB2016_train_dict.npy', train_dict)
```
